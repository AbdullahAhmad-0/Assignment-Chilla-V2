# 3- Rename dataframe columns?

```python
df = pd.DataFrame({'Column 1': [1, 2, 3, 4, 5], 'Column 2': [6, 7, 8, 9, 10]}) # make sure size of all column are same
df
```

```python
# 1 Rename using rename method
df.rename(columns={'Column 1': 'A', 'Column 2': 'B'}, inplace=True) # use inplace = true if you want to commit dataframe
df
```

```python
# 2 rename using columns method
df.columns=['B_', 'A_']
df
```

```python
# 3 to replace any character, string
df.columns = df.columns.str.replace('_', '')
df
```

```python
# 4 Adding prefix to columns
df = df.add_prefix('Start ')
df
```

```python
# 5 Adding suffx to columns
df = df.add_suffix(' End')
df
```
