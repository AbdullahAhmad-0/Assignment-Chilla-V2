# 6- Reverse Row order

```python
import seaborn as sns
import pandas as pd

# load dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# 1 reverse dataset
df.loc[::-1].head()
```

```python
# 2 reverse dataset also reset serial no of dataset
df.loc[::-1].reset_index(drop=True).head()
```

# 7- Reverse column order

```python
# 1 Reverse column order
df.loc[:, ::-1].head()
```

# 8- select a column by dtype

```python
# 1 find data types of all columns
df.dtypes
```

```python
# 1 only select those have numeric type
df.select_dtypes(include=['number']).head()
```

```python
# 2 only select those have object type
df.select_dtypes(include=['object']).head()
```

```python
# 3 only select those have bool type
df.select_dtypes(include=['bool']).head()
```

```python
# 4 select multiple data types
df.select_dtypes(include=['object', 'bool', 'number']).head()
```

```python
# 4 excluse multiple data types
df.select_dtypes(exclude=['object', 'bool', 'number']).head()
```

# 9- Convert strings to number

```python
df = pd.DataFrame({'col A': ['1.2', '2', '3','6','7','8','9','22'], 
                    'col B': ['4', '5', '6','7','8','9','10','11']})
df
```

```python
df.dtypes
```

```python
# 1 Convert object(string) to int
df.astype({'col A': 'float64', 'col B': 'int64'}).dtypes
```

```python
# 2 another way
pd.to_numeric(df['col A'], errors='coerce') # error='coerce' It will replace all non-numeric values with NaN
```
